## TASK 13: Counts of International flights

```
# 13a. What is the total number of international flights that flew to the United States in each of the 6 years
```

```
inUSADF = flightsAllDF.where((flightsAllDF.origin!="United States")&(flightsAllDF.dest=="United States")).groupby(flightsAllDF["year"]).sum("numflights").select("Year",F.col("sum(numflights)").alias("totalflightsin"))
inUSADF.sort(inUSADF["totalflightsin"].asc()).show()
```

```
# 13b. In what years were there more international flights flying to the United States than international flights that flew out of the United States
```

```
outUSADF = flightsAllDF.where((flightsAllDF.origin=="United States")&(flightsAllDF.dest!="United States")).groupby(flightsAllDF["year"]).sum("numflights").select("year",F.col("sum(numflights)").alias("totalflightsout")).sort(F.col("year").asc())
inoutUSADF = inUSADF.join(outUSADF,on="year",how="inner")
inoutUSADF.select('*',(inoutUSADF.totalflightsin > inoutUSADF.totalflightsout).alias("morein")).show()
```

## TASK 14: Aggregate statistics

```
# 14a. For the rows that contain United States as destination but NOT as origin (i.e. international flights into the USA) compute and print the following summary statistics for the column
```

```
flightsAllDF.where((flightsAllDF.origin!="United States")&(flightsAllDF.dest=="United States")).groupby(flightsAllDF["year"]).agg(F.sum("numflights").alias("sumflightstoUS"),F.round(F.avg("numflights"),2).alias("avgflightstoUS"),F.max("numflights").alias("maxflightstoUS")).sort(F.col("year").asc()).show()
```
