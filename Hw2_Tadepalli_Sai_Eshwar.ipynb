```python
# import statements
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# magic function
%matplotlib inline
```

## Task 2a - Read in an image as both PNG and JPG

```python
# read in the png and jpg
img_png = plt.imread("img1.png")
img_jpg = plt.imread("img1.jpg")
```

## Task 2b - Display images in the Notebook

```python
# write a display function and call it twice for each type of image
def display_image(img):
    plt.xlabel('columns')
    plt.ylabel('rows')
    plt.imshow(img)
    plt.colorbar()
    plt.show()
display_image(img_png)
display_image(img_jpg)
```

## Task 2c - Resizing the Figure

```python
# resize the png image and display it
plt.figure(figsize=(18,12))
display_image(img_png)
```

## Task 9 - Render the PNG image with the lower half of the image set at transparency level = 0.4. The rest of the image (the top half) should be at its original level

## Task 13 - Apply Gamma correction to the PNG image

```python
# Display two  gamma correctied images - light and dark
img_png_light = img_png[:,:,:3] ** (1/5)
display_image(img_png_light)
img_png_dark = img_png[:,:,:3] ** (5)
display_image(img_png_dark)
```

## Task 14 - Obtain and print one image formed by horizontally stacking 3 images – redscale, greenscale and bluescale of the original PNG image

```python
# obtain the img in red, green and blue scale and horizontally stack/ join them 
img_png_Red = img_png.copy()
img_png_Green = img_png.copy()
img_png_Blue = img_png.copy()
img_png_Red[:,:,[1,2]] = 0
img_png_Blue[:,:,[0,2]] = 0
img_png_Green[:,:,[0,1]] = 0
img_png_stacked = np.concatenate((img_png_Red, img_png_Blue, img_png_Green),axis=1)
plt.figure(figsize=(10,6))
display_image(img_png_stacked)
```
