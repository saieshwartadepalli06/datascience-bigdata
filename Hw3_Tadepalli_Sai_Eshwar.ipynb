## Task 11b - More complex operations on data for which built-in functions may not exist – using DF.apply()

```python
def expstandardize(x):
    return np.exp((x - x.mean())/x.std())
popDF[['y2019']].apply(expstandardize).head()
popDF[['y2018','y2019']].apply(expstandardize).head()
```

## Question 11.ii - Write code to compute a new column in popDF that contains the coefficient of variation for each state for the population in years 2010-2019. Use DF.apply() for this task. Name this column ‘coeffofvar’. Display the first 5 values of only this new column (with index).

### ANSWER 11.ii 

```python
def coeffvar(x):
    return (x.std()/x.mean())
popDF['covar'] = popDF.loc[:,'y2010':'y2019'].apply(coeffvar, axis = 1)
popDF['covar'].head()
```

## Task 11c - Aggregation operations using DF.agg()

```python
popDF[['y2019','y2018']].agg(['count','sum','min','max','mean','var','skew','kurt','median','idxmax','idxmin'])
popDF[['y2019','y2018']].agg({'y2019':['min','max'], 'y2018':['var','std']})
popDF[['y2019','y2018']].agg({'y2019':['min','max'], 'y2018':['var','std', np.min, np.max]})
```
